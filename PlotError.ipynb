```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
ABS_PATH = os.path.abspath("")
CSV_PATH = os.path.join(ABS_PATH, "CSV")


csvs = {fname: pd.read_csv(os.path.join(CSV_PATH, fname)) for fname in os.listdir(CSV_PATH)}
```

```python
date = "2023-11-24"
df = csvs[date]
df = df[7:]
```

```python

# Set dark theme
# Assuming your DataFrame is named 'df'
# Convert 'Date' column to datetime format
df['Datetime'] = pd.to_datetime(df['Date'], format='%H-%M-%S')

# Set custom dark mode colors
background_color = '#1E1E1E'
grid_color = '#2E2E2E'
text_color = '#FFFFFF'
line_color = sns.color_palette("rocket", 1)[-1]

# Plotting
plt.figure(figsize=(10, 6), facecolor=background_color)
plt.plot(df['Datetime'], df['Error'], marker='o', linestyle='-', markersize=8, color=line_color)
plt.title(f'{date}', color=text_color)
plt.xlabel('Time', color=text_color)
plt.ylabel('Error (Float)', color=text_color)

# Format x-axis to display only (hour, minute, second)
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%H:%M:%S'))

# Set grid and ticks color
plt.grid(True, color=grid_color)
plt.tick_params(axis='x', colors=text_color)
plt.tick_params(axis='y', colors=text_color)

# Set background color
plt.gca().set_facecolor(background_color)

# Show the plot
plt.show()
```
