```python
import numpy as np
import os
from scipy import stats
import matplotlib.pyplot as plt
```

```python
l = np.append(np.arange(100), np.arange(10))
```

$\displaystyle \sigma^2 = \frac{1}{N}\sum_{n=1}^N \left(u_n-u_x\right)^2$

$\displaystyle \mu_i = \frac{1}{N}\sum_{n=1}^N \left(u_n-u_x\right)^i$

```python
def central_moments(*, k, errors):

    errors = np.array(errors)
    errors_x = errors.mean()
    if k==1: return errors_x
    central_moment = ((errors-errors_x)**k).mean()
    return central_moment
```

```python

print(central_moments(k=1, errors=l))
print(l.mean(), "\n")

print(central_moments(k=2, errors=l)/central_moments(k=2, errors=l)**(0.5*0))
print(l.var(), "\n")

print(central_moments(k=3, errors=l)/central_moments(k=2, errors=l)**(0.5*3))
print(stats.skew(l), "\n")

print(central_moments(k=4, errors=l)/central_moments(k=2, errors=l)**(0.5*4))
print(stats.kurtosis(l, fisher=False))
```

```python
plt.plot(l)
```

$\displaystyle \frac{1}{\sigma\sqrt{2\pi}} \exp{\left(-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2\right)}$

```python
def get_distribution(*, x, mu, std):
    return (1/std*(2*np.pi)**0.5) * np.exp(-0.5*((x-mu)/std)**2)
```

```python
a = np.linspace(-100,300,100)
mu = central_moments(k=1, errors=l)
b = get_distribution(x=a, mu=mu, var=central_moments(k=2, errors=l))

plt.plot(b)
```
